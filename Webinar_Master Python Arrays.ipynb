### Array in Python

- **`Python Array`** is a collection of common type of data structures having elements with same data type
- It is used to store collections of data
- Arrays are handled by the "array" module
- Elements of the array must be of the same numeric type

        Python arrays are used when you need to use many variables which are of the same type
        It can also be used to store a collection of data
        The arrays are especially useful when you have to process the data dynamically
        Python arrays are much faster than list as it uses less memory

```python
# Syntax to create an array
# arrayName = array.array(type code for data type, [array,items])
```

1. `Identifier`: specify a name like usually, you do for variables
2. `Module`: Python has a special module for creating array in Python, called "array" â€“ you must import it before using it
3. `Method`: the array module has a method for initializing the array. It takes two arguments, type code, and elements.
4. `Type Code`: specify the data type using the type codes available (see list below)
5. `Elements`: specify the array elements within the square brackets, for example [130,450,103]

```python
# Calling the library for array
# !pip install array   # installing a library
import array    # calling a library


# Building an array using the above syntax
balance = array.array('i', [3,9,6,5,20,13,19,22,30,25])
len(balance)
type(balance)
balance
```

```python
# Declaring an list - this is not an array
students = ['Alex', 'Bill', 'Catherine', 'Andy', 'Molly', 'Rose']
students
type(students)
```

**1.1 Accessing elements inside array**

`The syntax is arrayName[indexNum]`

```python
min(balance)
max(balance)
```

```python
balance
print("Array first element is:",balance[0]) 
print("Array second last element is:",balance[-2])

# Multiple items inside array
print(balance[1:5])
balance[0:]
print(balance[7:10])
```

*`This is also known as slicing operation`*

**1.2 Insert elements in array**

Python array insert operation enables you to insert one or more items into an array at the beginning, end, or any given index of the array. This method expects two arguments index and value.

`The syntax is  arrayName.insert(index, value)`

```python
balance.insert(2, 145)
len(balance)
balance
```

**1.3 Modify elements in array**

Arrays in Python are mutable.

`The syntax is Object_name[index]=value`

```python
balance[0]=99 
balance
```

**1.4 Combinig two arrays**

```python
balance1 = array.array('b', [4, 6, 8]) 
balance2 = array.array('b', [9, 12, 15]) 

balance_result = array.array('b')   
balance_result = balance1 + balance2 
print(balance_result)
```

**1.5 Pop an element in array**

pop() method is used to pop(drop) an element from Python array
del could also be used to do the same thing

```python
balance
len(balance)

balance.pop(2) 
print(balance)
len(balance)

del balance[4]  
print(balance)
len(balance)
```

**1.6 Delete elements from array**

We can delete one item from an array by value using the remove method
- This method accepts only one argument, value
- After running this method, the array items are re-arranged, and indices are re-assigned

```python
balance
len(balance)

balance.remove(99)
len(balance)
balance
```

**1.7 Index value in array**

We can search for an item in an array based on its value
- This method accepts only one argument, value
- It is a non-destructive method, which means it does not affect the array values

`The syntax is arrayName.index(value)` 

```python
balance = array.array('i', [3,9,6,5,20,13,19,22,30,25])
balance.index(19)
```

**1.8 Reverse an array**

This operation will reverse the entire array

`Syntax is array.reverse()`

```python
balance.reverse()           
print(balance)
```

**1.9 Count occurence elements in array**

Count the occurrence of elements in the array using the array.count(x)

```python
balance = array.array('b', [2, 3, 5, 4,3,3,3]) 
print(balance.count(3))
```

**1.10 Traversing an array**

```python
for x in balance:
	print(x)
```

```python
# Reading array using list comprehension. Note that the output is list not an array
out = [elements for elements in balance]
type(out)
out
```

```python
# Task as requested by participant
names = ['edureka', 'notebook']
names.reverse()
print(names)

for ele in names:
    print([char for char in ele])
    print(type(ele))
```

```python
# Alternate way to reverse elements in list
rev_split_list = [[char for char in ele] for ele in names]
rev_split_list
type(rev_split_list)
# rev_split_list[0][::-1] + rev_split_list[1][::-1]
```

```python
# Reverse elements of the reversed list using 'map' function
' '.join(map(str, rev_split_list))

# Alternate way : reverse elements using list comprehension
' '.join([str(elem) for elem in rev_split_list])
```
